```python
import numpy as np
a=np.array([10,20,30])
print(a)
```

```python
num=np.array([i for i in range(1,21)])
print(num)
print("sum:",num.sum())
print("Standard deviation:",num.std())
print("Average:", num.sum()/len(num))
print("mean:",num.mean())
```

```python
matrix=np.array([[1,4,5],[4,6,7],[5,6,7]])
print(matrix)
row_sum=np.sum(matrix, axis=0)
print("ROW WISE SUM:", row_sum)
```
