```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset('iris')
#duplicating 10 rows

df=pd.concat([df, df.iloc[:10]], ignore_index=True)
#seed for Reproducibility

np.random.seed(42)
#Introduce missing in sepal_length(~15^%)
missing_sepal=np.random.choice(df.index,size=int(0.15*len(df)), replace=False)
df.loc[missing_sepal,'Sepal_length']=np.nan

#introduce missing in 'petal_width'(~10%)
missing_petal=np.random.choice(df.index,size=int(0.10*len(df)), replace= False)
df.loc[missing_petal, 'petal_width']=np.nan
print(df)
a=df["sepal_length"].mean()
a=df["sepal_length"].fillna(a, inplace=True)
print(a)
df.isnull().sum()



```
