```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree

# ---------------------------------------------------
# 1. Create the dataset (from uploaded file)
# ---------------------------------------------------
data = {
    'Age': ['Youth', 'Youth', 'Middle-aged', 'Senior', 'Senior', 'Senior',
            'Middle-aged', 'Youth', 'Youth', 'Senior', 'Youth',
            'Middle-aged', 'Middle-aged', 'Senior'],
    'Income': ['High', 'High', 'High', 'Medium', 'Low', 'Low',
               'Low', 'Medium', 'Low', 'Medium', 'Medium',
               'Medium', 'High', 'Medium'],
    'Student': ['No', 'No', 'No', 'No', 'Yes', 'Yes',
                'Yes', 'No', 'Yes', 'Yes', 'Yes',
                'No', 'Yes', 'No'],
    'Credit_rating': ['Fair', 'Excellent', 'Fair', 'Fair', 'Fair', 'Excellent',
                      'Excellent', 'Fair', 'Fair', 'Fair', 'Excellent',
                      'Excellent', 'Fair', 'Excellent'],
    'Buy_computer': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No',
                     'Yes', 'No', 'Yes', 'Yes', 'Yes',
                     'Yes', 'Yes', 'No']
}

df = pd.DataFrame(data)

# ---------------------------------------------------
# 2. Encode categorical variables
# ---------------------------------------------------
encoder = LabelEncoder()

for column in df.columns:
    df[column] = encoder.fit_transform(df[column])

# ---------------------------------------------------
# 3. Features and target
# ---------------------------------------------------
X = df[['Age', 'Income', 'Student', 'Credit_rating']]
y = df['Buy_computer']

# ---------------------------------------------------
# 4. Train Decision Tree (Entropy / Information Gain)
# ---------------------------------------------------
model = DecisionTreeClassifier(
    criterion='entropy',
    max_depth=3,
    random_state=42
)

model.fit(X, y)

# ---------------------------------------------------
# 5. Draw / Plot the Decision Tree
# ---------------------------------------------------
plt.figure(figsize=(20,10))
plot_tree(
    model,
    feature_names=X.columns,
    class_names=['No', 'Yes'],
    filled=True,
    rounded=True
)
plt.title("Decision Tree  Buy Computer Dataset")
plt.show()
```
