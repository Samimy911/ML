```python
mean=df["age"].mean()#find the mean of the missing value
df["age"].fillna(mean,inplace=True)# replacing it
df.isnull().sum()

```

```python
mean=df["deck"].mode()#find the mean of the missing value
df["deck"].fillna(mean,inplace=True)# replacing it
df.isnull().sum()
```

```python
value=df["embarked"].mode()[0]
df["embarked"].fillna(value, inplace=True)
df.isnull().sum()
df.groupby('age')['fare'].mean()

```
