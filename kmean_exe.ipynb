```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from mpl_toolkits.mplot3d import Axes3D
```

```python
data={
      
     'Feature1':[5.1,4.9,np.nan, 4.6,5.0,5.4,4.6,5.0,4.4,4.9],
     'Feature2':[3.5,3.0,3.2,3.1,np.nan,3.9,3.4,3.4,2.9,3.1],
     'Feature3':[1.4,1.4,1.3,np.nan,1.5,1.7,1.4,1.5,1.4,1.5]


}

df=pd.DataFrame(data)

print(data)
```

```python
df["Feature1"].fillna(df["Feature1"].mean(), inplace=True)
df["Feature2"].fillna(df["Feature2"].mean(), inplace=True) 
df["Feature3"].fillna(df["Feature3"].mean(), inplace=True)
```

```python
kmeans=KMeans(n_clusters=2,random_state=0)
kmeans.fit(df)
```

```python
centroids=kmeans.cluster_centers_
labels=kmeans.labels_
print("Centrois\n",centroids)
print("Labels",labels)
data_array=df.values
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
colors=['Red' if label==0 else 'blue' for label in labels]
ax.scatter(data_array[:,0], data_array[:,1],data_array[:,2], c=colors)
ax.scatter(centroids[:,0], centroids[:,1], centroids[:,2],s=300,c='Yellow',marker='*')

ax.set_xlabel("Feature1")
ax.set_ylabel("Feature2")
ax.set_zlabel("Feature3") 
ax.set_title("3D KMeans Clustering Example") 
plt.show()
```
