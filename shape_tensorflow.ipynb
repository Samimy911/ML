```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
import cv2 # Import OpenCV
import matplotlib.pyplot as plt # Import for displaying images

def generate_shape_image(size=28, shape_type='Circle'):
  """Generate a simple image with a shape (Circle or Square)."""
  # Fix: Corrected numpy.zeros syntax
  img = np.zeros((size, size), dtype=np.uint8)

  if shape_type == 'Circle':
    center = (size // 2, size // 2)
    radius = (size // 2) - 10
    cv2.circle(img, center, radius, 255, -1)
  elif shape_type == 'Square':
    # Add logic for drawing a square
    start_point = (size // 4, size // 4)
    end_point = (size - size // 4, size - size // 4)
    cv2.rectangle(img, start_point, end_point, 255, -1)
  else:
    print(f"Warning: Unknown shape_type '{shape_type}'. Generating empty image.")

  return img

# --- Demonstrate the function ---
print("Generating a circle image:")
circle_image = generate_shape_image(size=64, shape_type='Circle')
print("Circle image shape:", circle_image.shape)

plt.figure(figsize=(4, 4))
plt.imshow(circle_image, cmap='gray')
plt.title('Generated Circle')
plt.axis('off')
plt.show()

print("\nGenerating a square image:")
square_image = generate_shape_image(size=64, shape_type='Square')
print("Square image shape:", square_image.shape)

plt.figure(figsize=(4, 4))
plt.imshow(square_image, cmap='gray')
plt.title('Generated Square')
plt.axis('off')
plt.show()
```
